```python
# Pick 5 random samples This is my initial hypothesis picking 5 random samples to look at
samples = my_data.sample(5)

# All 16 categories
categories = [
    'GENDER', 'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
    'PEER_PRESSURE', 'CHRONIC DISEASE', 'WHEEZING', 'ALCOHOL CONSUMING',
    'SHORTNESS OF BREATH', 'COUGHING', 'SWALLOWING DIFFICULTY',
    'CHEST PAIN', 'LUNG_CANCER'
]
# For my output these are the only categories that I want
categoriesinsurvey = [
    'AGE', 'SMOKING', 'YELLOW_FINGERS', 'ANXIETY',
    'CHRONIC DISEASE', 'WHEEZING', 'ALCOHOL CONSUMING',
    'SHORTNESS OF BREATH', 'COUGHING', 'SWALLOWING DIFFICULTY',
    'CHEST PAIN'
]
# Displaying the categories that I do not want, as gender, peer pressure, and lung cancer is not indicative for a predictive model
exclude_from_pie_and_print = ['GENDER', 'PEER_PRESSURE','LUNG_CANCER']
# I want to have a for loop again, indicated by for and idx
for idx, row in samples.iterrows():
    yes_count = (row[categoriesinsurvey] == 2).sum()  # when the count is 2 it means Yes; reminder that 1 means no

    print("SAMPLE ID:", idx)
    print("GENDER:", row['GENDER']) # Since I removed it from the original yes/no sample survey to print out high/low risk but i still want to see the gender
    print("Yes count:", (yes_count), "/13") # there are 13 categories since we removed 3 from the original 16 categories

    if yes_count > 6: # greater than because that is over half of 16
        print("High risk for lung cancer ")
    else:
        print("Low risk for lung cancer ")
    print("-" * 40) #separates the 5 random samples from one another and i wanted to find a tool where i can visually see the difference between samples
```

```python
#I wanted to use the same samples from above and forloop so that it all happens in one cell
for idx, row in samples.iterrows():
    # Again I changed my nos and yesses to values where 1= No,2= Yes
    responses = row[categoriesinsurvey].replace({1: 'No', 2: 'Yes'})

    # I want to differentiate the yeses and nos with colors so I will have: Yes = Red and No = Blue
    colors = ['Red' if r == 'Yes' else 'blue' for r in responses] #originially i had accidentally had the opposite, so that was a big error originally for stakeholders

    plt.figure(figsize=(6, 6))
    plt.pie(
        [1]*len(responses),
        labels=responses.index, #this will ensure that the slides are equal
        colors=colors, # I wanted the two colors to be in my pie chart
        wedgeprops={'edgecolor': 'white', 'linewidth': 2}, #this separates the categories to be well seen
        autopct=lambda pct: '' , startangle=90) #placing the numbers and percentages; with lambda no numbers will be put on becsuse that is not important for my stakeholder

    plt.title(f"Sample ID: {idx} - Yes/No by Category") #the index for the variables that i am looking at which are yes and no for the categories
    plt.show()
```

```python
#I want to create a visual with all of the categories separated by gender to see the frequency per dr h request

# First i only want to isolate lung cancer
lung_cancer = my_data[my_data['LUNG_CANCER'].apply(lambda x: str(x).strip().upper() in ['2','YES'])].copy() #the strip makes sure all ways that yes was recorded is included

# Survey categories not including gender because i want to separate it by gender in two
categoriesinsurvey = [
    'AGE','SMOKING','YELLOW_FINGERS','ANXIETY',
    'PEER_PRESSURE','CHRONIC DISEASE','WHEEZING',
    'ALCOHOL CONSUMING','SHORTNESS OF BREATH','COUGHING',
    'SWALLOWING DIFFICULTY','CHEST PAIN','LUNG_CANCER'
]

# I must convert the yes and no values into 0 and 1
for col in categoriesinsurvey:
    lung_cancer[col] = lung_cancer[col].apply(lambda x: 1 if str(x).strip().upper() in ['2','YES'] else 0).astype(int)

# I am comparing the two genders
male = lung_cancer[lung_cancer['GENDER']=='M']
female = lung_cancer[lung_cancer['GENDER']=='F']

# I want to display by frequency
male_frequency = male[categoriesinsurvey].mean()
female_frequency = female[categoriesinsurvey].mean()

# DataFrame
frequency_df = pd.DataFrame({'Male': male_frequency, 'Female': female_frequency})

# Drop columns that are all NaN
frequency_df = frequency_df.dropna(how='all')

# Heatmap
plt.figure(figsize=(10,6))
sns.heatmap(frequency_df, annot=True, cmap='Reds', cbar_kws={'label':'Proportion with Yes'})
plt.title("Common Survey Factors in Lung Cancer Patients by Gender")
plt.ylabel("Survey Factor")
plt.show()
```
